# Demonstrating the effect of axis tilt correction

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1000,1001)
tilt=np.array([0.05,0.1,0.2,0.5,1])

def line_cross_talk(x,tilt) :
    return x*np.tan(np.pi*tilt/180)

for th in tilt :
    plt.plot(x,line_cross_talk(x,th), label=r'$\theta$={0:0.2f}'.format(th))

plt.fill_between(x,y1=0,y2=1,color='green',alpha=0.2)
plt.fill_between(x,y1=1,y2=2.5,color='yellow',alpha=0.2)
plt.fill_between(x,y1=2.5,y2=line_cross_talk(x[-1],tilt[-1]),color='red',alpha=0.2)
plt.legend()
plt.grid('on')
plt.gca().set_xticks(np.linspace(0,1000,11))
plt.axis('tight')
plt.tight_layout()
plt.xlabel('Distance from center of rotation')
plt.ylabel('Number of projection lines from ideal')
plt.savefig('tilt_angles.png',dpi=150)
plt.savefig('tilt_angles.svg')
```
