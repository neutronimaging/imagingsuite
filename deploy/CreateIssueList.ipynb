```python
import numpy as np
import pandas as pd
```

```python
issues1=pd.read_json("https://api.github.com/repos/neutronimaging/imagingsuite/issues?state=closed&page=1&since=\"2023-10-30T10:00:00Z\"&per_page=400")
issues2=pd.read_json("https://api.github.com/repos/neutronimaging/imagingsuite/issues?state=closed&page=2&since=\"2023-10-30T10:00:00Z\"&per_page=400")
issues3=pd.read_json("https://api.github.com/repos/neutronimaging/imagingsuite/issues?state=closed&page=3&since=\"2023-10-30T10:00:00Z\"&per_page=400")
#issues4=pd.read_json("https://api.github.com/repos/neutronimaging/imagingsuite/issues?state=closed&page=4&since=\"2019-06-01T10:00:00Z\"&per_page=400")
#issues5=pd.read_json("https://api.github.com/repos/neutronimaging/imagingsuite/issues?state=closed&page=5&since=\"2019-06-01T10:00:00Z\"&per_page=400")
frames = [issues1, issues2, issues3]

issues = pd.concat(frames)
```

```python
issues.count()
```

```python
closedissues=issues[(issues['state']=="closed") & (issues['pull_request'].isnull())]
closedissues[['number','title','closed_at']]
```

# Print list for github markdown

```python
for index,issue in closedissues.iterrows() :
    print('- Issue',issue['number'], '- [',issue['title'],'](', issue['html_url'],')')
```
